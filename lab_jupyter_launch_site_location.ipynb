```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
from math import radians, cos, sin, asin, sqrt

# Calcular distancia entre dos puntos usando la fórmula de Haversine
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Radio de la Tierra en km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    return R * c
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )
# Coordenadas del sitio de lanzamiento CCAFS LC-40
launch_lat = 28.5623
launch_lon = -80.5774

# Coordenadas del punto costero más cercano (puedes tomarlo de MousePosition o elegir manualmente)
coast_lat = 28.56367
coast_lon = -80.57163

# Calcular distancia
distance = calculate_distance(launch_lat, launch_lon, coast_lat, coast_lon)
from folium.features import DivIcon

# Crear un nuevo mapa centrado en el sitio
coast_map = folium.Map(location=[launch_lat, launch_lon], zoom_start=12)

# Marcar sitio de lanzamiento
folium.Marker(
    location=[launch_lat, launch_lon],
    popup='Launch Site: CCAFS LC-40',
    icon=folium.Icon(color='blue')
).add_to(coast_map)

# Marcar punto costero
folium.Marker(
    location=[coast_lat, coast_lon],
    popup='Closest Coastline',
    icon=folium.Icon(color='green')
).add_to(coast_map)

# Mostrar la distancia con un marcador de texto
distance_marker = folium.Marker(
    location=[(launch_lat + coast_lat)/2, (launch_lon + coast_lon)/2],  # Punto medio
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance),
    )
)
distance_marker.add_to(coast_map)

# Mostrar línea conectando ambos puntos (opcional)
folium.PolyLine(locations=[[launch_lat, launch_lon], [coast_lat, coast_lon]], color='gray').add_to(coast_map)

# Mostrar el mapa
coast_map
```

```python
import folium
from folium.features import DivIcon
from math import radians, cos, sin, asin, sqrt

# Función para calcular distancia entre dos puntos
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    return R * c

# Coordenadas del sitio de lanzamiento y del punto de interés (e.g. ciudad)
launch_coord = [28.5623, -80.5774]       # CCAFS LC-40
poi_coord = [28.6100, -80.8070]          # Ejemplo: ciudad o infraestructura

# Calcular distancia
distance = calculate_distance(*launch_coord, *poi_coord)

# Crear mapa centrado en el sitio
site_map = folium.Map(location=launch_coord, zoom_start=10)

# Agregar marcador del sitio de lanzamiento
folium.Marker(
    launch_coord,
    popup='Launch Site',
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Agregar marcador del punto cercano (ciudad/carretera/ferrocarril)
folium.Marker(
    poi_coord,
    popup='Closest City/Rail/Highway',
    icon=folium.Icon(color='purple')
).add_to(site_map)

# Agregar marcador de distancia en el punto medio
midpoint = [(launch_coord[0] + poi_coord[0]) / 2, (launch_coord[1] + poi_coord[1]) / 2]
folium.Marker(
    midpoint,
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance)
    )
).add_to(site_map)

# Dibujar línea entre el sitio y el punto cercano
folium.PolyLine([launch_coord, poi_coord], color='gray', weight=2).add_to(site_map)

# Mostrar el mapa
site_map
from IPython.display import display

display(site_map)
```
